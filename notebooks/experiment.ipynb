```python
import numpy as np
from ftdrf.simulator import simulate
import matplotlib.pyplot as plt

psi0 = np.array([1.0, 0.0])
psi1 = np.array([0.0, 1.0])
params = {
    'kappa': 0.5, 'E_n': 1.0, 'E_nm1': 0.8, 'F45': 0.1, 'F44': 1.0, 'd_hat': np.array([1.0,1.0])
}
seq = simulate(psi0, psi1, params, steps=20)
norms = [np.linalg.norm(s) for s in seq]
plt.plot(norms, marker='o')
plt.xlabel('n')
plt.ylabel('||psi||')
plt.title('FTDRF sample run')
plt.grid(True)
plt.show()
```
